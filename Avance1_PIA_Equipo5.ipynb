* **El objetivo es una clasificación con los datos, asi que hay que acomodar los datos de manera que se nos facilite esta tarea.**

Antes de seguir avanzando, es conveniente filtrar las columnas de la tabla con las que estaremos trabajando, y desechar las que no, ya que contamos con un gran número de las mismas. 
### **Columnas que serán eliminadas:**
* **Row ID:** Solo es un contador de las filas, para ello ya contamos con los que proporciona el propio Python al guardar la base de datos.
* **Player NR y PID:** Al tener "Player ID", esta columna puede ser desechada.
* **Birthdate:** Esto debido a que al contar con la columna "AGE" pasa a ser irrelevante para nuestro estudio.
* **Team short name:** Al igual que el caso anterior, contamos con las columnas "TEAM" y "TEAM ID", además, hay muchos equipos que tienen valores nulos en este apartado.
* **Potential:** Nos enfocamos en lo que ya pasó, no en lo que pudo haber pasado, este apartado no nos aprota nada.
* **Atacking y Defending Work Rate:** Estas columnas no representan la calidad del jugador en dichos aspectos, es más que nada sobre los esfuerzos que realiza, y tampoco es nuestro interés, ya que perseguimos únicamente los resultados.
* **Possesion Count:** La posesion no es un aspecto importante para calificar a un jugador.
* **Match API ID:** No será relevante ni útil para nosotros esta columna.

```python
#Eliminamos las columnas
del(base_soccer['ROWID'])
del(base_soccer['PID'])
del(base_soccer['PLAYERNR'])
del(base_soccer['BIRTHDATE'])
del(base_soccer['TEAM_SHORT_NAME'])
del(base_soccer['POTENTIAL'])
del(base_soccer['DEFENSIVE_WORK_RATE'])
del(base_soccer['ATTACKING_WORK_RATE'])
del(base_soccer['POSSESIONCOUNT'])
del(base_soccer['MATCH_API_ID'])
#El orden de las columnas esta perfecto tal cuál están en la base de datos, por lo que no modificaremos dicho apartado.
```

### 2. Introducción general

La Bolsa de Nueva York (NYSE) es el mayor mercado de valores del mundo en volumen monetario y el primero en número de empresas adscritas. Su volumen en acciones fue superado por la del NASDAQ en los 1990, pero el capital de las compañías listadas en la NYSE es 5 veces mayor que en el NASDAQ. La Bolsa de Nueva York cuenta con un volumen anual de transacciones de 36 billones de dólares, incluidos los 12 billones de compañías no estadounidenses.

Impresionante, ¿verdad? Cada día se hacen compras y ventas de valores, y esta demás el decir para qué sería útil una herramienta de este tipo, por ejemplo, con los precios históricos de las acciones, se puede realizar un análisis para hallar el momento ideal para entrar o salir de la empresa, entre muchas cosas más...

### 3. Descripción de los datos

**a) Tipo de datos**  
Se tienen 4 archivos que contienen las bases de datos.  
En la primera tenemos una tabla, donde se encuentran todos los fundamentos necesarios para procesar la información, tales como las métricas anuales contenidas en el periodo de 2012 a 2016.  
La segunda se trata de una tabla como "raw data" que contiene los precios establecidos pero sin procesar.
La tercera está basada en la segunda, ya que se trata de una tabla que contiene almacenados los precios ajustados.  
Y la última, se trata de una tabla con información general de los sectores de cada empresa.
  
**b) Descripción de las columnas**  
En la primera base de datos tenemos una tabla compuesta por 79 columnas, donde el principal ordenamiento es por fechas. Las demás columnas son de datos y métricas a lo largo de los años. Dichas métricas varían entre totales, valores promedios, valores anuales, etc.  
En la segunda base tenemos una tabla compuesta por 7 columnas, en donde tenemos una columna de fecha y el resto de columnas son los precios establecidos en cada año, tales como el precio de apertura, cierre, bajo, alto, etc.  
En la tercera tenemos la misma información que en la segunda, sin embargo se realizó un ajuste que servirá en el despliegue de dicha información.
Por último, en la cuarta base tenemos una tabla de 8 columnas, de las cuales tenemos información importante con los nombres y sectores de cada empresa, así como las regiones y áreas que ayudarán a asignar a las empresas por el factor que se requiera en ese momento.
