* **Investigar los siguientes comandos: ?df.dropna() ?df.astype**

En cuanto al comando **df.dropna** lo podemos llamar como DataFrame.dropna, este comando nos ayuda a eliminar los datos nulos que tenemos en una base de datos. Este comando tiene la siguiente sintaxis con sus respectivos parámetros df.dropna(axis,how,thresh,subset,inplace). Estos parámetros pueden contener los siguientes valores


*   **Axis:** Puede tomar el valor de 0(o index) y 1(o columns), por default toma el valor de 0. El valor que le damos a la función en este parámetro le indica a la función que eliminar cuando encuentre un valor nulo, si le damos un 0 la función eliminirá la fila donde se encuentre el valor nulo, por el contrario, si es 1 la función eliminará la columna donde este el valor nulo    
*   **How:** Puede tomar el valor de "any" o "all". Este parámetro es el criterio de eliminación de una fila o columa, si decidimos poner "any" como valor de este parámetro el comando eliminirá la fila o columna que contenga al menos un valor nulo, en cambio, si el valor del parámetro "how" es "all" el comando solo eliminará la fila o columna cuando este llena de datos nulos


*   **Thresh:** Este parámetro debe ser un valor entero y es opcional. Se le podriá ver como un seguro, el número entero que pondremos aquí es el minimo de datos NO nulos que debe contener una fila o columna para salvarse de ser eliminada 
*   **Subset:** Es un array opcional. Este parámetro contiene el nombre de unas filas o columnas que queremeos que sean revisadas por el comando. Por ejemplo, si el valor de axis es 1 aqui pondremos el nombre de las filas que queremos que tambien sean consideradas a revisar por el comando

*   **Inplace:** Valor boleano que por default es False. Si el valor de este parámetro es True devuelve un valor nulo al realizar el comando  

Fuente: https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.dropna.html

Por su parte la función **df.astype()** cambia el tipo de datos de los objetos a uno en especifico, nos devuelve el dataframe con los tipos de datos fundidos. Su sintaxis es df.astype(dtype,copy,errors,kwargs). Estos parámetros pueden tener los siguientes valores



*   **dtype:** Es el tipo de dato que le queremos asignar a nuestro objeto

*   **Copy:** Valor boleano que por default es True. Cuando es True nos devuelve una copia del objeto modificado. Hay que ser cuidadoso al poner un False en este parámetro ya que los valores cambiados pueden propagarse a otros objetos

*   **Errors:** Puede tener el valor de raise o ignore. Este parámetro controla el lanzamiento de excepciones sobre datos no válidos para el tipo de dato proporcionado. Con raise ignora las excepciones, y con ignore las elimina, si ocurre un error al momento de correr la función regresa el objeto original
*   **kwargs:** Palabra clave que sirve de argumento para pasar al constructor

Fuente: https://www.geeksforgeeks.org/python-pandas-dataframe-astype/
